# GSE59856 - PANCREATIC CANCER

```python
pc = pd.read_csv('/Users/chaitanyasharma/Desktop/BIOADA/GSE59856/B00013_TopN_200_fact.csv')
pc.head()
```

```python
pc.shape
```

```python
pc.describe()
```

```python
pc.isnull().sum()
```

```python
pc.drop(['age','Random','RandomBin','gender','tumor_stage','geo_accession','operability','target'],axis=1,inplace=True)
pc.head(10)
```

```python
from sklearn.utils import shuffle
pc = shuffle(pc)
pc.head()
```

```python
pc.isnull().sum()
```

```python
pc.shape
```

```python
pc.dropna(inplace=True)
pc.shape
```

```python
y = pc['target_factor']
X = pc.drop(['target_factor'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,roc_curve
```

```python
clf = RandomForestClassifier()#,min_samples_leaf =1,min_samples_split=2,oob_score=False,bootstrap=True)
Rf = clf.fit(X_train, y_train)
```

```python
y_pred = Rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
params = {
            'n_estimators': [10, 50, 100,200],
            'max_depth': [1, 3,4,5],
            'bootstrap': [True, False],
            'max_features': ['auto', 'sqrt'],
            'min_samples_leaf': [1, 2,3,4,5],
            'min_samples_split': [2, 3,4,5],
            'oob_score': [True, False]
        }

grid = GridSearchCV(estimator=RandomForestClassifier(), param_grid=params, cv=3, verbose=2)
randommodel=grid.fit(X_train, y_train)
```

```python
rf = randommodel.best_estimator_
print(f'CV score: {randommodel.best_score_}')
for key, value in randommodel.best_params_.items():
    print(f'{key}: {value}')
```

```python
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
